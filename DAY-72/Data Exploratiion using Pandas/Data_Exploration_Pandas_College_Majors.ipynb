```python
clean_df = college_major_dataset.dropna()
```

```python
clean_df.tail()
```

```python
clean_df['Starting Median Salary']
```

```python
# Find the Max Salary
clean_df['Starting Median Salary'].max()
```

```python
# Finding the index of the max salary
index_for_max_salary = clean_df['Starting Median Salary'].idxmax()
print(index_for_max_salary)
```

```python
# Finding the major with the highest salary
clean_df['Undergraduate Major'].loc[index_for_max_salary]
```

```python
#   Finding which  major has the highest mid career level salary
idx_max_mid_career_level = clean_df['Mid-Career Median Salary'].idxmax()
clean_df['Undergraduate Major'].loc[idx_max_mid_career_level]
```

```python
# which college major has lowest starting salary
idx_min_starting_salary = clean_df['Starting Median Salary'].idxmin()
clean_df['Undergraduate Major'].loc[idx_min_starting_salary]
```

```python
# how much do the graduates earn after the university
clean_df['Mid-Career Median Salary'].loc[idx_min_starting_salary]
```

```python
# which college major has lowest mid career level salary
idx_min_mid_career_salary = clean_df['Mid-Career Median Salary'].idxmin()
clean_df['Undergraduate Major'].loc[idx_min_mid_career_salary]
```

```python
# how much people are expect to earn with this degree
clean_df['Starting Median Salary'].loc[idx_min_mid_career_salary]
```

```python
spread_column = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_column)
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# degrees with highest potential
highest_potential_df = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential_df[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head(5)
```

```python
great_spread_df = clean_df.sort_values('Spread', ascending=False)
great_spread_df[['Undergraduate Major', 'Spread']].head(5)
```

```python
# Grouping the data using group by

clean_df.groupby('Group').count()
```

```python
clean_df.groupby('Group').mean()
```

```python
numeric_columns = ['Spread', 'Starting Median Salary', 'Mid-Career Median Salary', 'Mid-Career 10th Percentile Salary', 'Mid-Career 90th Percentile Salary']
pd.options.display.float_format = '{:,.2f}'.format
# calculate the mean for all the numeric columns
clean_df.groupby('Group')[numeric_columns].mean()
```
